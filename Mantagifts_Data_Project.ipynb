```python
import pandas as pd
import numpy as np

df_customers = pd.read_excel('Manta customers.xlsx')
df_product_sales = pd.read_excel('Manta product sales per month.xlsx')
df_products = pd.read_excel('Manta products.xlsx')

display(df_customers.head())
display(df_product_sales.head())
display(df_products.head())
```

```python
# Display the shape of each dataframe
print("Shape of df_customers:", df_customers.shape)
print("Shape of df_product_sales:", df_product_sales.shape)
print("Shape of df_products:", df_products.shape)

# Display the column names and data types of each dataframe
print("\nInfo of df_customers:")
df_customers.info()

print("\nInfo of df_product_sales:")
df_product_sales.info()

print("\nInfo of df_products:")
df_products.info()

# Display basic descriptive statistics for numerical columns
print("\nDescription of df_customers:")
display(df_customers.describe())

print("\nDescription of df_product_sales:")
display(df_product_sales.describe())

print("\nDescription of df_products:")
display(df_products.describe())
```

```python
print("=== Product Sales dataset ===")
print("Vorm (rows, cols):", df_product_sales.shape)
print("\nEerste 5 rijen:\n", df_product_sales.head())
print("\nInformatie over kolommen:\n")
print(df_product_sales.info())

print("\n--- Analyses ---")
print("Totaal aantal verkopen:", df_product_sales['sales_amount'].sum())
print("\nVerkopen per maand:\n", df_product_sales.groupby('month')['sales_amount'].sum())
print("\nTop 10 producten (op basis van omzet):\n",
      df_product_sales.groupby('product_id')['sales_amount'].sum().sort_values(ascending=False).head(10))
print("\nAantal missende waarden per kolom:\n", df_product_sales.isnull().sum())

```
